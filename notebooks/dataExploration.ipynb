```python
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
# Show sample images
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
real_imgs = os.listdir('data/real_vs_fake/real-vs-fake/train/real')[:5]
fake_imgs = os.listdir('data/real_vs_fake/real-vs-fake/train/fake')[:5]

for i, fname in enumerate(real_imgs):
    img = Image.open(f'data/real_vs_fake/real-vs-fake/train/real/{fname}')
    axes[0,i].imshow(img)
    axes[0,i].set_title('REAL')
    axes[0,i].axis('off')

for i, fname in enumerate(fake_imgs):
    img = Image.open(f'data/real_vs_fake/real-vs-fake/train/fake/{fname}')
    axes[1,i].imshow(img)
    axes[1,i].set_title('FAKE')
    axes[1,i].axis('off')

plt.tight_layout()
plt.savefig('sample_images.png')
plt.show()
```
