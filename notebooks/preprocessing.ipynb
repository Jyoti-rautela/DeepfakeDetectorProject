```python
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
```

```python
# These exact values are required for ALL pre-trained models
MEAN = [0.485, 0.456, 0.406]
STD = [0.229, 0.224, 0.225]

train_transforms = transforms.Compose([
    transforms.Resize((300, 300)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(MEAN, STD)
])

val_transforms = transforms.Compose([
    transforms.Resize((300, 300)),
    transforms.ToTensor(),
    transforms.Normalize(MEAN, STD)
])

# Auto-loads images from folder — 0=fake, 1=real
train_data = ImageFolder('../data/real_vs_fake/real-vs-fake/train', transform=train_transforms)
valid_data = ImageFolder('../data/real_vs_fake/real-vs-fake/valid', transform=val_transforms)
test_data  = ImageFolder('../data/real_vs_fake/real-vs-fake/test',  transform=val_transforms)

# Load in batches of 32
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=False)
test_loader  = DataLoader(test_data,  batch_size=32, shuffle=True)

print(f'Train : {len(train_data)} images')
print(f'Val   : {len(valid_data)} images')
print(f'Test  : {len(test_data)}  images')
print(f'Classes: {train_data.class_to_idx}')
```
